```python
# normalizar customer, phone, internet y account
customer_normal = pd.json_normalize(datos['customer'])
phone_normal = pd.json_normalize(datos['phone'])
internet_normal = pd.json_normalize(datos['internet'])
account_normal = pd.json_normalize(datos['account'])

```

```python
datos = datos.drop(columns=['customer']).join(customer_normal)
datos = datos.drop(columns=['phone']).join(phone_normal)
datos = datos.drop(columns=['internet']).join(internet_normal)
datos = datos.drop(columns=['account']).join(account_normal)
datos
```

```python
# Verifique si hay valores √∫nicos en cada columna
for col in datos.columns:
  print(f"valores √∫nicos en la columna'{col}': {datos[col].nunique()}")
  if datos[col].nunique() < 50: # Imprima los valores solamente si no hay muchos
    print(datos[col].unique())
    print('-' * 50)  # L√≠nea divisoria para mejor visualizaci√≥n
```
